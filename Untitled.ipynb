```python
import pandas as pd
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt


df = pd.read_csv('Updated_GPS_11_1.CSV', names=['LATITUDE', 'LONGITUDE','HEADING'], sep=',', skiprows = 1)
```

```python

known_point = [51.529111,-0.059452] #middleton green
#known_point = [51.530588, -0.066351] # Ion square gardens
Distances =[]
for i in range(len(df)):
    NM = np.arccos(np.sin(df.iloc[i]['LATITUDE']*np.pi/180)*np.sin(known_point[0]*np.pi/180)+np.cos(df.iloc[i]['LATITUDE']*np.pi/180)*np.cos(known_point[0]*np.pi/180)*np.cos(known_point[1]*np.pi/180-df.iloc[i]['LONGITUDE']*np.pi/180))*3443.8985
    Distances.append(NM*1852)
```

```python
Sorted_dist = np.sort(Distances)
print(Sorted_dist)
count = []
for idx in range(len(df)):
    count.append(idx)
```

```python
plt.scatter(Sorted_dist,count)
```

```python
# Fit a normal distribution to the data:
mu, std = norm.fit(Distances)
print(mu)
print(std)
```

```python
plt.xlim(-5,15)

p = norm.pdf(count, mu, std)
plt.plot(count, p)
plt.hist(Distances,15, density=True)
title = "Fit results: mu = %.2f,  std = %.2f" % (mu, std)
plt.title(title)

```
