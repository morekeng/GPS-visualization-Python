```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import folium


df = pd.read_csv('Updated_GPS_17.CSV', names=['LATITUDE', 'LONGITUDE','HEADING'], sep=',', skiprows = 1)
```

```python
m = folium.Map([51.53, -0.067], tiles="OpenStreetMap", zoom_start=16)
kw = {"prefix": "fa", "color": "blue", "icon": "arrow-up"}

for i in range(0,len(df)):
    angle= int(df.iloc[i]['HEADING'])
    icon = folium.Icon(angle=angle, **kw)
    
    folium.map.Marker(location= [df.iloc[i]['LATITUDE'], df.iloc[i]['LONGITUDE']],
                icon=icon).add_to(m)

```
